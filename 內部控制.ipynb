# PCA 降維

```python
pca = PCA(n_components=3)
pca_df_tfidf = pca.fit_transform(df_tfidf)
pca_df_tfidf
```

# k-means 分群

```python
LABEL_COLOR_MAP = {'內控':'pink' , '內控1' : 'cyan'}
label_color = [LABEL_COLOR_MAP[l] for l in text_label] 
plt.figure(figsize = (25,6))
plt.scatter(pca_df_tfidf[:,0],pca_df_tfidf[:,2], c= label_color, alpha=1)
plt.legend()
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3)

#Compute cluster centers and predict cluster indices
tfidf_clustered = kmeans.fit_predict(pca_df_tfidf)

LABEL_COLOR_MAP = {0:'r', 1: 'g', 2: 'b'}
label_color = [LABEL_COLOR_MAP[l] for l in tfidf_clustered]

plt.figure(figsize = (4,4))
plt.scatter(pca_df_tfidf[:,0],pca_df_tfidf[:,2], c= label_color, alpha=0.5) 
plt.show()
```
